```python
countyID = {'Bronx':'005',
           'Manhattan':'061',
           'Staten Island':'085',
           'Brooklyn':'047',
           'Queens':'081'}
```

```python
ct.columns
ct['boro_code'].unique(), ct['boro_name'].unique(), ct['boro_ct_2010'].unique()
#bronx = (ct[['boro_name']] == 'Bronx').values.flatten()
#bronx, ct[['boro_code']].values[bronx]
```

```python
ct['allcodes'] = list(zip(ct.boro_name, ct.ct_2010))
# Create Point Geometry for based on lonlat column
#ct['GEOID'] = ct.allcodes.map(lambda x:int("36%03d%06d"%(int(x[0]), int(x[1]))))
#type(ct['GEOID'].values[0]), type(demo['GEOID'].values[0])
ct['GEOID']  = ct[['allcodes']].applymap(lambda x: 
                                         int("36%03d%06d"%(int(countyID[x[0]]),int(x[1]))))
```

```python
ct['GEOID'].head()
```

```python
merged2010 = gpd.GeoDataFrame(pd.merge(ct, demo, on='GEOID'))
merged2010.plot()
merged2010.columns
```

```python
import sklearn as skl
vals = skl.preprocessing.scale(merged2010[['TotalPopulation',
       'PopulationDensity', 'AreaLand', 'MedianAge', 'MinorityPercentage',
       'UnemploymentPercentage', 'MedianHouseholdIncome',
       'HouseholdGiniIndex']].dropna().values)
```

```python
nc = 5

km = skl.cluster.KMeans(n_clusters=nc)


km.fit(vals)

clustersdf = gpd.GeoDataFrame()
clustersdf['GEOID'] = merged2010[['GEOID','TotalPopulation',
       'PopulationDensity', 'AreaLand', 'MedianAge', 'MinorityPercentage',
       'UnemploymentPercentage', 'MedianHouseholdIncome',
       'HouseholdGiniIndex']].dropna()['GEOID']
clustersdf['cluster'] = km.labels_
```

```python
import pylab as pl
ax = pl.figure(figsize=(15,15)).add_subplot(111)
clustersdf.merge(merged2010, on="GEOID", how='inner').plot(column='cluster', colormap='Blues', alpha = 1, axes=ax)
ax.axis('off')
```
